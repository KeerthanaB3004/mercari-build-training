```python
import pprint
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

```

```python
# Load Wine dataset
wine_data = datasets.load_wine()
```

```python
# Problem 1-1: Extract and output only data from the dataset
wine_features = wine_data.data
print("Extracted Data:\n", wine_features)
```

```python
# Problem 1-2: Convert data into a pandas DataFrame
wine_df = pd.DataFrame(wine_features, columns=wine_data.feature_names)
```

```python
# Problem 1-3: Show the first 10 rows
print(wine_df.head(10))
```

```python
# Problem 1-4: Check number of rows and columns
print("Shape of wine_df:", wine_df.shape)
```

```python
# Problem 1-5: Check the list of column names
print("Column names:", wine_df.columns.tolist())
```

```python
# Problem 1-6: Extract the 'alcohol' column
print(wine_df['alcohol'])
```

```python
# Problem 1-7: Extract 'color_intensity' and 'hue' columns
print(wine_df[['color_intensity', 'hue']])
```

```python
# Problem 2-1: Extract rows with magnesium > 100
high_magnesium_df = wine_df[wine_df['magnesium'] > 100]
print(high_magnesium_df)
print("Rows with magnesium > 100:", high_magnesium_df.shape[0])
```

```python
# Problem 2-2: Sort by 'ash' in descending order
print(wine_df.sort_values(by='ash', ascending=False).head(10))
```

```python
# Problem 2-3: Group by 'target' and calculate mean 'flavanoids'
wine_df['target'] = wine_data.target
grouped_flavanoids = wine_df.groupby('target')['flavanoids'].mean()
print(grouped_flavanoids)
print("Target with smallest avg flavanoids:", grouped_flavanoids.idxmin())
```

```python
# Problem 2-4: Create is_alcohol_high function
def is_alcohol_high(alcohol):
    return 1 if alcohol >= 13 else 0

wine_df['is_alcohol_high'] = wine_df['alcohol'].apply(is_alcohol_high)
print(wine_df.tail(10))
```

```python
# Problem 3-1: Describe statistics
print(wine_df.describe())
print("Feature with largest standard deviation:", wine_df.std().idxmax())
```

```python
# Problem 3-2: Histogram of 'proline'
wine_df['proline'].plot.hist(bins=30, alpha=0.7, color='purple')
plt.xlabel('Proline')
plt.title('Histogram of Proline')
plt.show()
```

```python
# Problem 3-3: Scatter plot 'total_phenols' vs 'flavanoids'
wine_df.plot.scatter(x='total_phenols', y='flavanoids', alpha=0.7, color='blue')
plt.title('Total Phenols vs Flavanoids')
plt.show()
```

```python
# Problem 3-4: Compute correlation coefficient
total_phenols_correlation = wine_df[['total_phenols', 'flavanoids']].corr()
print(total_phenols_correlation)
```

```python
# Problem 4-1: Calculate mean for each wine type
mean_values = wine_df.groupby('target').mean()
print(mean_values)
selected_features = mean_values.std().sort_values(ascending=False).index[:2]
print("Selected features for classification:", selected_features)
```

```python
# Problem 4-2: Scatter plot of selected features
sns.scatterplot(data=wine_df, x=selected_features[0], y=selected_features[1], hue='target', palette='coolwarm')
plt.title(f"Scatter Plot of {selected_features[0]} vs {selected_features[1]}")
plt.show()
```
